# Maps
 a function that takes one set of values and maps them to another set of values. In data science we often have a need for creating new representations from existing data, or for transforming data from the format it is in now to the format that we want it to be in later.



There are two mapping methods that you will use often.
1. map ()
2. apply ()

````python
# take a pandas seires and apply a function i each element in this seires
mapping = {'male': 0, 'female': 1}
````

````python
data['sex_numeric'] = data['Sex'].map(mapping)
data
````

````python
data['fare_category'] = data['Fare'].map(lambda price: 'High' if price > 50 else 'Standard')
data
````

apply() is the equivalent method if we want to transform a whole DataFrame by calling a custom method on each row.
- Element-wise, row-wise, or column-wise
- axis=0: Apply function to each column.
- axis=1: Apply function to each row.

````python
def age_group(age):
    if pd.isna(age):
        return "Unknown"
    elif age < 18:
        return "Child"
    else:
        return "Adult"

data["age_group"] = data["Age"].apply(age_group)
data.age_group
````

````python
data
````

````python
data["person_info"] = data.apply(lambda row: f"{row['Sex']} - {row['Age']}", axis=1)
data.person_info
````

# group by

Split-Apply-Combine:
1. Split: The data is split into groups based on one or more keys.
2. Apply: A function is applied to each group.function like sum(), mean(), count(). but it can also be a transformation or a filtering function.
3. Combine: The results of the function applications are combined into a new DataFrame or Series.

```
df.groupby(by=grouping_columns)[columns_to_show].function()
```

- mean(): Calculates the average of the values.
- sum(): Calculates the sum of the values.
- count(): Counts the number of non-null values.
- size(): Counts the total number of values (including nulls).
- min(): Finds the minimum value.
- max(): Finds the maximum value.
- std(): Calculates the standard deviation.
- nunique(): Counts the number of unique values.

"What was the survival rate for males versus females?"


````python
data_s = data.groupby("Sex")["Survived"].mean()
data_s
````

````python
data_pclass= data.groupby("Pclass")["Survived"].count()
data_pclass
````

#.agg

````python
# Group by both 'Pclass' and 'Sex'
# Calculate the mean of 'Survived' (survival rate) and 'Age'
detailed_analysis = data.groupby(['Pclass', 'Sex']).agg(
    survival_rate=('Survived', 'mean'),
    average_age=('Age', 'mean'),
)
detailed_analysis
````

# fillna

Replacing missing values is a common operation. Pandas provides a really handy method for this problem: fillna(). fillna() provides a few different strategies for mitigating such data. For example, we can simply replace each NaN with an "Unknown":



````python
data
````

````python
data.Cabin.fillna("Unknown",inplace=True)
````

````python
data
````

````python
data.fillna({
    'Age': data['Age'].mean(),
    'Fare': data['Fare'].median()
}, inplace=True)
````

````python
data.Age.fillna(method='ffill', inplace=True) # Fills missing values with the value from the previous row
````

````python
data.fillna(method='bfill', inplace=True) # Fills missing values with the next available value.
````
